# Indeskleme ve Dilimleme
* pandas kütüphanesi özelinde nasıl old. öğreniyoruz

```python
import pandas as pd

dictionary = {"isim": ["Ali", "Veli","kenan", "murat", "ayse", "hilal","mehmet", "ismail", "merve", "yusuf", "selin", "tugay", "serkan", "serra", "ozen", "mehmetcan"],
             "yas": [15,16,17,33,45,66, 23, 34, 19, 20, 21, 22, 23, 24, 25, 26],
             "maas": [1000, 2000, 3000, 4000, 5000, 6000, 7000, 8000, 9000, 10000, 11000, 12000, 13000, 14000, 15000, 16000]}

data = pd.DataFrame(dictionary)
data
```

```python
# yas sütununu alalım

data["yas"]
```

```python
# veriye yeni bir sütun ekleyelim.

data["sehirler"] = ["Ankara", "Istanbul", "Izmir","Giresun","Antalya", "Aydın", "Kocaeli", "Bursa", "Manisa", "Edirne", "Van", "Mersin", "Samsun", "Malatya", "Erzurum", "Artvin"]

data
```

__loc[] ve icoc[] yöntemleri:__ indeskleme yapılırken kullanılır.

__loc[]__ location(konum)

__icoc[]__ index location (indeks konum)

```python
# verinin yas sütununu baska bir yöntem ile alalım.

data.loc[:,"yas"]
```

```python
# yas sütunu ve index3'e kadar olan satırları al. (Not 3 dahil)
data.loc[:3, "yas"]
```

```python
#indeksi 3e kadar olan satırları ve yas'tan sehirlere kadar olan sütunları al.

data.loc[:3,"yas":"sehirler"]
```

```python
#indeksi 3e kadar olan  satırlar ile yas ve isim sütunlarını al

data.loc[:3,["isim","yas"]]
```

```python
#satırları tersten yazdır

data.loc[::-1,:]
```

```python
# :, 1 tüm satırları getir ve indeksi 1 olan sütun der.

#yas sütunu iloc ile yazdır, yas 1 nolu indekse sahip

data.iloc[:,1]
```

```python
# ilk 3 satırı ve yas ile isim sütunlarını al iloc ile

data.iloc[:3,[0,1]]
```

# Filtreleme

* Veri analizinde filtreleme en çok kullanılan yöntemlerden biridir. Veriyi anlamlandırmak için gruplara ayırmak gerekir, gruplara ayırma işlemini filtreleme yöntemi ile yaparız.

```python
import pandas as pd

dictionary = {"name": ["Alice", "Bob", "Charlie", "Charlotte", "Julia", "Selin", "Alex"],
              "age": [25, 30, 35, 20, 45, 26, 28],
              "country": ["USA", "USA", "UK", "Australia", "USA", "Turkey", "Germany"]
             }
             
veri = pd.DataFrame(dictionary)
veri
```

```python
# ilk olarak age-yas göre filtreleme yapıyoruz.

filtre1 = veri.age > 22
```

```python
filtrelenmis_veri = veri[filtre1]

filtrelenmis_veri
```

```python
# simdi yası 22den büyük ve USAde yasayan örnekleri alalım.

filtre2 = veri.country == "USA"
veri[filtre1 & filtre2]
```

## Liste Kavramı

* Veriyi işlemek için kullanılan etkili bir yöntemdir.

```python
import pandas as pd

sozluk = {"isim": ["Ali", "Veli","kenan", "murat", "ayse", "hilal","mehmet", "ismail", "merve", "yusuf", "selin", "tugay", "serkan", "serra", "ozen", "mehmetcan"],
             "yas": [15,16,17,33,45,66, 23, 34, 19, 20, 21, 22, 23, 24, 25, 26],
             "sehirler": ["Ankara", "Istanbul", "Izmir","Giresun","Antalya", "Aydın", "Kocaeli", "Bursa", "Manisa", "Edirne", "Van", "Mersin", "Samsun", "Malatya", "Erzurum", "Artvin"],
             }
data = pd.DataFrame(sozluk)
data
```

```python
# ortalama yası bulalım

ortalama_yas = data.yas.mean()

ortalama_yas
```

* Ort yası bulduktan sonra verimizi büyük küçük oü ikiye ayıralım.
* Ort. yasları büyük yasa sahip = büyük, diğer örn kücük olarak etiketleyelim

```python
data["YAS_GRUBU"] = [ "kucuk" if ortalama_yas > i else  "büyük" for i in data.yas]

data
```

```python
data.columns = [i.lower() for i in data.columns]
data
```

**lambda fonksiyonu** liste kavramı yerine kullanılır.

```python
data = data.assign(yas_5_yil_sonra = lambda x: (x['yas'] + 5))
data
```

**apply metodunu** liste kavrama ve lambda fonk yerine kullanabiliriz.

```python
def yas10YilSonra(age):
    return age + 10
```

```python
data["yas_10_yil_sonra"] = data.yas.apply(yas10YilSonra)

data
```

## Birleştirme

```python
import pandas as pd

sozluk1 = {"isim": [ "murat", "ayse", "hilal"],
          "yas": [33, 45, 66],
          "sehirler": ["Giresun", "Antalya", "Aydın"]}

veri1 = pd.DataFrame(sozluk1)
veri1
```

```python
sozluk2 = {"isimler": ["selin", "tugay", "serkan"],
          "yaslar": [20, 21, 22],
          "sehirler": ["Van", "Mersin", "Samsun"]}

veri2 = pd.DataFrame(sozluk2)
veri2
```

```python
#dikey birleştirme

veri_dikey =pd.concat([veri1, veri2], axis = 0)

veri_dikey
```

```python
#yatay birleştirme

veri_yatay =pd.concat([veri1, veri2], axis = 1)

veri_yatay
```
