## SAÇILIM GRAFİĞİ

* Saçılım diyagramı iki farklı değişkenin arasındaki ilişkiyi incelemek için kullanılır.
* Aralarındaki ilişkinin sebebi görülmese de, ilgili iki değişken arasında direkt olarak ilişki bulunup bulunmadığı ve bu ilişkinin ne derece güçlü olduğu görülebilir.

```python
#Kütüphanelerimizi içe aktaralım.

import matplotlib.pyplot as plt
import pandas as pd

veri = pd.read_csv("veri_setleri/Iris.csv")
veri.head()
```

```python
# Alt yaprak uzunluğunu --> sepal -length  alt yaprak genişliğine sepal width göre çizdirelim

plt.figure()
plt.scatter(veri["SepalLengthCm"], veri["SepalWidthCm"], color="magenta", s = 15, alpha = 0.7) #saçılım old. scatter kullandık | s = size , noktaların boyutu alpha saydamlık
plt.title('Alt Yaprak Uzunluğu ve Genişlikleri')
plt.xlabel("Alt yaprak uzunlugu (Cm)")
plt.ylabel("Aly yaprak genişiği (Cm)")
plt.grid(True)
plt.show()

    
```

**İlişki keşfedilemedi fakat 3 cm genişliğinde alt yaprağı olan çiçeklerin, alt yaprak uzunluğunun çeşitliliğini farkedebiliriz.**

## HİSTOGRAM GRAFİĞİ

* **HİSTOGRAM:** gruplandırılmış verilerin bir sütun grafiğinde gösterilmesidir.
* Diğer bir ifadeyle; tekrarlı sayılardan olusan verilerin, uygç işlemlerden sonra önce tabloya, tablodan yararlanılarak grafiklere aktarılmasıdır. Yani veri gruplarının grafiğinin dikdörtgen sütunlar halinde gösterilmesidir.

```python
#Kütüphanelerimizi içe aktaralım.

import matplotlib.pyplot as plt
import pandas as pd

veri = pd.read_csv("veri_setleri/Iris.csv")
veri.head()
```

```python
# Sepal length (alt yaprak uzulugu) ve petal length (üst yaprak uzunlugu) göre çizdirilmesi 


plt.figure()
plt.hist(veri["SepalLengthCm"], color = "hotpink", alpha=0.5, bins = 20, label = "Alt yaprak uzunlugu") #bins histogramdaki aralıklar için kullanılır
plt.hist(veri["PetalLengthCm"], color = "cyan", alpha = 0.5, bins = 20, label = "Ust Yaprak uzunlugu")
plt.title("Alt Yaprak Uzunlugu ve Ust Yaprak Uzunlugu")
plt.xlabel("Cm")
plt.ylabel("Frekans")
plt.legend() #labellerın gorunebilmesi icin

plt.show()


# tüm degerleri 20 esit parcaya bölüyor ve 20 esit parca içine düşenlerin sayısını veriyor.
```

## ÇUBUK GRAFİĞİ

* Kategorik verileri temsil ettikleri değerlerle orantılı olarak yükseklik-uzunluktaki dikdörtgen çubuklarla sunan bir grafiktir.

* Çubuklar dikey veya yatay olabilir.

```python
#Kütüphanelerimizi içe aktaralım.

import matplotlib.pyplot as plt
import pandas as pd

veri = pd.read_csv("veri_setleri/Iris.csv")
veri.head()
```

```python
veri_sutun_liste = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']
tanım = veri.describe()
tanım
```

```python
# SepalLengthCm	SepalWidthCm	PetalLengthCm	PetalWidthCm değerlerinin mean ortalamalarına ulasmak istiyoruz.

#iloc kullanırız index location

ortalama = tanım.iloc[1,[1,2,3,4]]
ortalama
```

```python
# alt-üst yaprak uzunluk ve genişliklerinin ortalamalarını bar plot ile çizdirelim.

plt.figure()
plt.bar(veri_sutun_liste,ortalama)
plt.title("Alt Yaprak Uzunluk ve Genişliklerinin Ortalamaları")
plt.ylabel("Cm")
plt.show()
```

## ALT PLAN GRAFİĞİ

* Alt plan grafiklerinde birden fazla grafik tek bir figür/çerçeve üzerinde bulunur.

```python
#Kütüphanelerimizi içe aktaralım.

import matplotlib.pyplot as plt
import pandas as pd

veri = pd.read_csv("veri_setleri/Iris.csv")
veri.head()
```

```python
fig, ax = plt.subplots(2,1) #figür üzerine 2*1 lik bir tablo olustur. bu tablo üzerine ax(eksen) ekle

fig.suptitle("Alt Yaprak Uzunluk ve Üst Yaprak Uzunluk", fontsize=14)
ax[0].plot(veri.Id, veri.SepalLengthCm, color = "red", label= "SepalLengthCm")
ax[0].set_title("Alt Yaprak Uzunluk") #baslık
ax[0].set_xlabel("id") #x ekseni label
ax[0].set_ylabel("Cm") #y ekseni label


ax[1].plot(veri.Id, veri.PetalLengthCm, color = "blue", label= "PetalLengthCm")
ax[1].set_title("Ust Yaprak Uzunluk") #baslık
ax[1].set_xlabel("id") #x ekseni label
ax[1].set_ylabel("Cm") #y ekseni label


fig.tight_layout() #figur arası yerlesım ayarla
fig.subplots_adjust(top=0.85)
plt.show()


```
